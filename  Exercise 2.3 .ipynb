### 4. Create a column with the count of trips

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
daily_trips = (
    df.groupby("date")
      .size()
      .reset_index(name="trip_count")
)
daily_trips.head()
```

### b. Merge trip counts with Weather Data

```python
weather_df = pd.read_csv(
    "/Users/ryumugil/Desktop/Achievement7/Citibike-dashboard-project/lga_weather_2022_daily.csv"
)

weather_df["date"] = pd.to_datetime(weather_df["date"])

weather_trips_df = weather_df.merge(
    daily_trips,
    on="date",
    how="left"
)
weather_trips_df.head()
```

```python
weather_trips_df[["date", "trip_count"]].head()
```

### 5. Create a line chart plotted on a dual axis.


```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(weather_trips_df["date"], weather_trips_df["trip_count"])
ax1.set_xlabel("Date")
ax1.set_ylabel("Bike trips per day")
ax1.set_title("Daily CitiBike Trip Counts and Temperature (2022)")

ax2 = ax1.twinx()
ax2.plot(weather_trips_df["date"], weather_trips_df["TMAX"])
ax2.plot(weather_trips_df["date"], weather_trips_df["TMIN"])
ax2.set_ylabel("Temperature (F)")

plt.show()
```

### 6. Explanation of the code
 
 Matplotlib was used to create a dual-axis line chart showing daily CitiBike trip counts and daily temperature for 2022. A figure and primary axis were first created using plt.subplots(). The bike trip counts were plotted on the primary (left) y-axis, and a secondary y-axis was created using ax.twinx() to plot the temperature data on the same time-based x-axis.

This visualization uses Matplotlibs object-oriented (OO) paradigm, where figure and axis objects (fig, ax1, and ax2) are explicitly created and controlled. This approach allows greater flexibility, such as adding a second y-axis and customizing each axis independently.
